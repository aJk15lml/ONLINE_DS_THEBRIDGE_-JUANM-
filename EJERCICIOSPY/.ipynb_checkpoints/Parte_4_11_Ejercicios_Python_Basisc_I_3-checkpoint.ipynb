## Ejercicio 1
Tenemos las siguientes variables:

```Python
A = 4
B = "Text"
C = 4.1
```

Comprueba:
1. Si A y B son equivalentes
2. Si A y C NO son equivalentes
3. Si A es mayor que C
4. Si C es menor o igual que A
5. Si B NO es equivalente a C

````python
A = 4 
B = "Text" 
C = 4.1
````

````python
print(A == B)  
print(A != C)
print(A > C)
print(C <= A)
print(B != C)
````

## Ejercicio 2
Crea un programa donde se recojan dos inputs del usuario, y el output del programa sea si esos inputs son iguales o no

````python
A = (input("¿Qué quieres que sea 'A'?: "))
B = (input("¿Y 'B'?: "))

print("'A' y 'B' son iguales, eso es: ", A == B)
````

## Ejercicio 3
Mismo programa que en 2, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

````python
A = (input("Dame el primer valor: "))
B = (input("Segundo valor por favor: "))
C = (input("Último valor: "))

print("Son todos iguales: ", A == B == C)
print("Al menos dos lo son: ", A == B or A == C or B == C)
````
